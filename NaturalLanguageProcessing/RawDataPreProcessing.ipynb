```python
import pandas as pd
```

```python
path = "../NaturalLanguageProcessing/Data_raw.xlsx"

df = pd.read_excel(path)
```

```python
df.head()
```

```python
df.drop(columns=['submitted_price', 'winning_status'], inplace=True)
```

```python
df.head(-10)
```

```python
df.columns
```

```python
list_picking = ['client_name','project_name', 'bid_classification', 'bid_type', 'bid_ID',
       'awarded_date', 'awarded_year']
```

```python
df.project_name
```

```python
df_new = df.groupby(list_picking)['company_name'].apply(lambda text: ''.join(text.to_string(index=False).strip())).str.replace('(\\n)', '').str.replace('\\t', '').str.strip().reset_index()
```
